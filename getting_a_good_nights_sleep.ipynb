### Which occupation has the lowest average sleep duration? 
Save this in a string variable called `lowest_sleep_occ`.

````python
display(sleep.groupby('Occupation')['Sleep Duration'].mean().sort_values())
lowest_sleep_occ = sleep.groupby('Occupation')['Sleep Duration'].mean().idxmin()
print(f'{lowest_sleep_occ = }')
````

### Which occupation has the lowest average sleep quality?
Save this in a string variable called `lowest_sleep_quality_occ`.

````python
display(sleep.groupby('Occupation')['Quality of Sleep'].mean().sort_values())
lowest_sleep_quality_occ = sleep.groupby('Occupation')['Quality of Sleep'].mean().idxmin()
print(f'{lowest_sleep_quality_occ = }')
````

### Did the occupation with the lowest sleep duration also have the lowest sleep quality?
If so assign a boolean value to variable `same_occ` variable, True if it is the same occupation, and False if it isn't.

````python
same_occ = (lowest_sleep_occ == lowest_sleep_quality_occ)
print(f'{same_occ = }')
````

### Let's explore how BMI Category can affect sleep disorder rates. 

Start by finding what ratio of app users in each BMI Category have been diagnosed with Insomnia. 

Create a dictionary named: `bmi_insomnia_ratios`. The key should be the `BMI Category` as a string, while the value should be the ratio of people in this category with insomnia as a float rounded to two decimal places. 

````python
def calculate_insomnia_ratio(df, bmi_category):
    """Calculate the ratio between app users with insomnia vs those that does for each bmi_category

    Args:
        df (dataframe): input dataframe containing the "BMI Category" and "Sleep Disorder" columnns.
        bmi_category (string): string value to compare against in the BMI Category column.

    Returns:
        float - Rounded ratio value.
    """
    filtered_df = df[(df["BMI Category"] == bmi_category) & (df["Sleep Disorder"] == "Insomnia")]
    total_rows = len(df[df["BMI Category"] == bmi_category])
    return round(len(filtered_df) / total_rows, 2)

# BMI categories
bmi_categories = sleep['BMI Category'].unique()

# Calculate insomnia ratios for each BMI category
bmi_insomnia_ratios = {category: calculate_insomnia_ratio(sleep, category) for category in bmi_categories}

print(f'{bmi_insomnia_ratios = }')
````
